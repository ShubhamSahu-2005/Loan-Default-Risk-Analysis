```python
import pandas as pd
from scipy.stats import chi2_contingency

df = pd.read_csv("../data/merged_loan_data.csv")


def chi_square_test(data, feature, target='defaulted'):
    """
    Performs Chi-Square Test of Independence.
    """

    print("=" * 70)
    print(f"CHI-SQUARE TEST: {feature} vs {target}")
    print("=" * 70)

    contingency = pd.crosstab(data[feature], data[target])

    print("\nContingency Table")
    print(contingency)

    chi2, p, dof, expected = chi2_contingency(contingency)

    print(f"\nChi-Square Statistic : {chi2:.4f}")
    print(f"Degrees of Freedom   : {dof}")
    print(f"P-value              : {p:.6f}")

    alpha = 0.05

    if p < alpha:
        print("\n Statistically Significant")
        print(f"{feature} and {target} are associated.")
    else:
        print("\n Not Statistically Significant")
        print(f"No evidence of association between {feature} and {target}.")

    return {
        "Feature": feature,
        "Chi2": round(chi2, 3),
        "P-value": round(p, 6),
        "Significant": p < alpha
    }
```

```python
results = []

results.append(chi_square_test(df, 'score_bucket'))
results.append(chi_square_test(df, 'dti_bucket'))
results.append(chi_square_test(df, 'home_ownership'))
results.append(chi_square_test(df, 'employment_status'))
results.append(chi_square_test(df, 'loan_purpose'))

results_df = pd.DataFrame(results)

print("\nSummary")
print(results_df)
```
