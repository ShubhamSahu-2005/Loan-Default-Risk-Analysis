## Data Cleaning — Loan Default Risk Analysis
**Input:** borrower_profiles.csv, loan_applications.csv  
**Output:** borrower_profiles_clean.csv, loan_applications_clean.csv  
**Steps:** Null handling → Type fixes → Outlier removal → Text standardization → Join key validation

```python
import pandas as pd 
borrowers=pd.read_csv(r'C:\Users\Shubham\Project\Loan Default Risk Analysis\data\borrower_profiles.csv')
loans=pd.read_csv(r'C:\Users\Shubham\Project\Loan Default Risk Analysis\data\loan_applications.csv')
```

```python
print("Borrowers Nulls")
print(borrowers.isnull().sum())

print("Loans Nulls")
print(loans.isnull().sum())
```

```python
print("----Borrowers Datatypes---")
print(borrowers.dtypes)

print("Loans DataTypes")
print(loans.dtypes)
```

```python
print("Borrowers duplicates",borrowers.duplicated().sum())
print("Loans Duplicates:",loans.duplicated().sum())
```

```python

print("=== BORROWER STATS ===")
print(borrowers[['age', 'annual_income', 'credit_score', 'years_employed', 
                  'existing_monthly_debt', 'dependents']].describe())

print("\n=== LOAN STATS ===")
print(loans[['loan_amount', 'interest_rate', 'dti_ratio', 
             'monthly_payment', 'days_delinquent']].describe())
```

```python
print("employment_status:", borrowers['employment_status'].unique())
print("home_ownership:", borrowers['home_ownership'].unique())
print("education_level:", borrowers['education_level'].unique())
print("loan_purpose:", loans['loan_purpose'].unique())
print("loan_status:", loans['loan_status'].unique())
print("defaulted:", loans['defaulted'].unique())
```

```python
borrower_id=set(borrowers['borrower_id'])
loan_id=set(loans['borrower_id'])

print("Id in borrowers but not in loans",len(borrower_id-loan_id))
print("Id in loans but not in borrowers",len(loan_id-borrower_id))
```

Clean files exported as `borrower_profiles_clean.csv` and `loan_applications_clean.csv`

```python

borrowers.to_csv('../data/borrower_profiles_clean.csv', index=False)
loans.to_csv('../data/loan_applications_clean.csv', index=False)

print("✅ Clean files exported successfully")
print(f"Borrowers: {len(borrowers)} rows")
print(f"Loans: {len(loans)} rows")
```
